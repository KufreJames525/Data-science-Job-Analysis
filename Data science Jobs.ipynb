### Visualizing Salary Trends Over Time

#### 1. Average Salary by Year

```python
data.head(2)
```

```python
avg_salary_year = data.groupby('work_year')['salary_in_usd'].mean().reset_index()
avg_salary_year
```

```python
plt.figure(figsize=(12,6), facecolor = "gray")

sns.lineplot(x = 'work_year', y = 'salary_in_usd', data = avg_salary_year )
plt.title("Average Salary (USD) in Data Science Jobs by Year")
plt.xlabel("Year")
plt.ylabel("Average Salary in USD")
plt.show()
```

#### Salary Distribution by Experience Level

```python
avg_salary_experience = data.groupby('experience_level')['salary_in_usd'].mean().reset_index()
avg_salary_experience
```

```python
plt.figure(figsize = (12,6))
sns.barplot(y='experience_level', x='salary_in_usd', data=avg_salary_experience, palette="viridis")
plt.title("Salary Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=data, palette="viridis")
plt.title("Salary Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.show()
```

#### Salary Based on Employment Type 

```python
data.head(2)
```

```python
plt.figure(figsize = (12,6))
sns.boxplot(x ='employment_type', y ='salary_in_usd', data = data , palette = 'Set2')
plt.title("Salary Distribution by Employment Type")
plt.xlabel("Employment Type")
plt.ylabel("Salary in USD")
plt.show()
```

#### Salary Distribution by Company Size

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='company_size', y='salary_in_usd', data=data, palette="coolwarm")
plt.title("Salary Distribution by Company Size")
plt.xlabel("Company Size")
plt.ylabel("Salary in USD")
plt.show()
```

#### Salary Distribution by Remote Level

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=data, palette="Blues")
plt.title("Salary Distribution by Remote Work Ratio")
plt.xlabel("Remote Ratio")
plt.ylabel("Salary in USD")
plt.show()
```

## Geographic Analysis

### Average Salary By Employee Residence

```python
avg_salary_by_residence = data.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).head(10).reset_index()
avg_salary_by_residence
```

```python
# Calculate average salary by employee residence
# avg_salary_by_residence = data.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).head(10).reset_index()

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(x= 'employee_residence', y = 'salary_in_usd', data = avg_salary_by_residence)
#avg_salary_by_residence.plot(kind='bar', color='coral')
plt.title("Top 10 Employee Residences by Average Salary (USD)")
plt.xlabel("Employee Residence")
plt.ylabel("Average Salary in USD")
plt.show()
```

#### Average Salary By Company location

```python
data.head(2)
```

```python
# Calculate average salary by company location

avg_salary_by_location = data.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(10).reset_index()
avg_salary_by_location
```

```python
# Plot
plt.figure(figsize=(12, 8))
#avg_salary_by_location.plot(kind='bar', color='seagreen')
sns.barplot(x= "company_location", y = 'salary_in_usd', data = avg_salary_by_location, color='gold')
plt.title("Top 10 Company Locations by Average Salary (USD)")
plt.xlabel("Company Location")
plt.ylabel("Average Salary in USD")
plt.show()
```

### Correlation Analysis

```python
numeric_data = data.select_dtypes(include = 'number')
numeric_data
```

```python
# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Plot heatmap
plt.figure(figsize=(12, 4))
sns.heatmap(correlation_matrix, annot=True ) #cmap="coolwarm" , vmin=-1, vmax=1)
plt.title("Correlation Matrix for Data Science jobs")
plt.show()
```

#### Working With Subplot

```python
plt.figure(figsize = (12, 8))
plt.subplot(2,2,1)
sns.heatmap(correlation_matrix, annot=True ) #cmap="coolwarm" , vmin=-1, vmax=1)
plt.title("Correlation Matrix for Data Science jobs")
plt.subplot(2,2,2)
sns.barplot(x= "company_location", y = 'salary_in_usd', data = avg_salary_by_location, color='gold')
plt.title("Top 10 Company Locations by Average Salary (USD)")
plt.xlabel("Company Location")
plt.ylabel("Average Salary in USD")
plt.subplot(2,2,3)
sns.barplot(x= 'employee_residence', y = 'salary_in_usd', data = avg_salary_by_residence)
#avg_salary_by_residence.plot(kind='bar', color='coral')
plt.title("Top 10 Employee Residences by Average Salary (USD)")
plt.xlabel("Employee Residence")
plt.ylabel("Average Salary in USD")
plt.subplot(2,2,4)
sns.boxplot(x='experience_level', y='salary_in_usd', data=data, palette="viridis")
plt.title("Salary Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.tight_layout()
plt.show()
```
